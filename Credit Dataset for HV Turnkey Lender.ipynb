# Business Objectives:
### The objective of this project is to identify factors that contribute to credit card defaults, which can help the credit card company to reduce losses, mitigate risks, and improve customer satisfaction. By gaining insights into the data and understanding the relationships between variables, the credit card company can make informed decisions to improve its business operations, marketing efforts, and customer targeting.

# Dataset:



### The columns/features in the given dataset are as follows:

### ID : ID of each client
### AMT : Amount of given credit in dollars (includes individual and family/supplementary credit
### GENDER : Gender (1=male, 2=female)
### EDUCATION : (1=graduate school, 2=university, 3=high school, 4=others, 5=unknown)
### MARITAL STATUS : Marital status (1=married, 2=single, 3=others)
### AGE : Age in years
### REPAY_SEP : Repayment status in September, 2005 (0=pay duly, 1=payment delay for one month, 2=payment delay for two months, … 8=payment delay for eight months, 9=payment delay for nine months and above)
### REPAY_AUG : Repayment status in August, 2005 (scale same as above)
### REPAY_JUL : Repayment status in July, 2005 (scale same as above)
### REPAY_JUN : Repayment status in June, 2005 (scale same as above)
### REPAY_MAY : Repayment status in May, 2005 (scale same as above)
### REPAY_APR : Repayment status in April, 2005 (scale same as above)
### AMTBILL_SEP : Amount of bill statement in September, 2005
### AMTBILL_AUG : Amount of bill statement in August, 2005
### AMTBILL_JUL : Amount of bill statement in July, 2005
### AMTBILL_JUN : Amount of bill statement in June, 2005
### AMTBILL_MAY : Amount of bill statement in May, 2005
### AMTBILL_APR : Amount of bill statement in April, 2005
### PRE_SEP : Amount of previous payment in September, 2005
### PRE_AUG : Amount of previous payment in August, 2005
### PRE_JUL : Amount of previous payment in July, 2005
### PRE_JUN : Amount of previous payment in June, 2005
### PRE_MAY : Amount of previous payment in May, 2005
### PRE_APR : Amount of previous payment in April, 2005
### DEF_AMT : Default payment (1=yes, 0=no)


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=[8,4]
```

```python
data=pd.read_csv('data set/Credit Dataset for HV Turnkey Lender.csv')
```

# Importing and cleaning data

```python
data.head()
```

```python
data.describe().T
```

```python
data.info()
```

```python
# Replacing Values of Gender Categories : 
# 1 = MAlE 
# 2 = FEMALE

data['GENDER']=data['GENDER'].replace(1,'Male')
data['GENDER']=data['GENDER'].replace(2,'Female')
```

```python
# Replacing Values of EDUCATION Categories : 
# 1 = Graduate School 
# 2 = University
# 3 = High School 
# 4 = Others
# 5 = Unknown

data['EDUCATION']=data['EDUCATION'].replace(1.0,'Graduate School')
data['EDUCATION']=data['EDUCATION'].replace(2.0,'University')
data['EDUCATION']=data['EDUCATION'].replace(3.0,'High School')
data['EDUCATION']=data['EDUCATION'].replace(4.0,'Others')
data['EDUCATION']=data['EDUCATION'].replace(5.0,'Unknown')
```

```python
# Replacing Values of MARITAL STATUS Categories :
# 1 = Married 
# 2 = Single
# 3 = Others 


data['MARITAL STATUS']=data['MARITAL STATUS'].replace(1,'Married')
data['MARITAL STATUS']=data['MARITAL STATUS'].replace(2,'Single')
data['MARITAL STATUS']=data['MARITAL STATUS'].replace(3,'Others')
```

```python
# We see that the EDUCATION column includes the categories (UNKNOWN) , 
#which are not included in the data specification. Assigning these values to (others) :

data.loc[data['EDUCATION'].isin(['Unknown']),'EDUCATION']='Others'

```

```python
# Calculation of Null Values In Data set:
print('Number of Null values in column:')
data.isna().sum()
```

```python
plt.figure(figsize=(4,6))
data.isna().sum().value_counts().plot(kind='pie',explode=[0.1,0],autopct="%1.1f%%")
plt.plot()
plt.title('Percentage of null Values in dataset')
```

```python
# There are 14 null values in  the EDUCATION Column :
data[data['EDUCATION'].isna()]['EDUCATION']
```

```python
# There are 14 null values in  the EDUCATION Column & we fill with value='Others' 
data.fillna(value='Others',inplace=True)
```

```python
# Again we Calculating Null Values In Data set:
print('Number of Null values in column:')
data.isna().sum()
```

```python
# We Calculting Duplicates rows
print('Number of duplicate rows:')
data.duplicated().sum()
```

# Exploratory Data Analysis (EDA)

```python
# How many % defaulters in dataset:
perc_default = data.DEF_AMT.sum() / len(data.DEF_AMT)
perc_default
```

```python
print('The percentage of defaulters in the data is',perc_default*100,'%')
data['DEF_AMT'].value_counts().plot(kind='pie',explode=[0.1,0],autopct="%1.1f%%")
plt.plot()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(data=data, x='AMT', kde=True, palette='viridis',hue=data.DEF_AMT)
plt.title('Distribution of (AMT)', fontsize=16)
plt.xlabel(' (AMT)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
print('It shows that the largest group of amount of credit limit is 40k to 50k')
```

```python
#Is there a connection between EDUCATION and defaulters?
sns.boxplot(data=data,x="GENDER",y='AMT')
plt.title('Relationship between GENDER and AMT', fontsize=16,color='BLUE')
plt.xlabel('GENDER',fontsize=12,color='GREEN')
plt.ylabel('AMT',fontsize=12,color='GREEN')
plt.show()
print('The credit limit is quite balanced between both the genders.')
```

```python
education=data[['EDUCATION',"DEF_AMT"]].groupby('EDUCATION',axis=0).sum().sort_values('DEF_AMT',ascending=False)
education
```

```python
#Is there a connection between EDUCATION and defaulters?
sns.heatmap(education, annot=True, fmt='d', cmap='YlGnBu', cbar=False)
plt.show()
print('People of graduate school, university, and high school are highly close to making deault payment.')
```

```python
#Is there a connection between marriage and defaulters?

education = data[['MARITAL STATUS', 'DEF_AMT']].groupby(['MARITAL STATUS'], axis=0).sum().sort_values('DEF_AMT', ascending=False)

plt.figure(figsize=(6,4))
plt.title('MARITAL STATUS & DEF_AMT',fontsize=16,color='BLUE')
plt.xlabel('DEF_AMT',fontsize=12,color='BLUE')
plt.ylabel('MARITAL STATUS',fontsize=12,color='BLUE')

sns.heatmap(education, annot=True, fmt='d', cmap='coolwarm', cbar=False, linewidths=0.5, linecolor='lightgray')
plt.show()
print('It shows that single people are most likely to default followed by married people. ')
```

```python
palette = "Set2"

# Plotting the relationship between variables
plt.figure(figsize=(8,6))
sns.countplot(x="MARITAL STATUS", hue="EDUCATION", data=data, palette=palette)
plt.title("Relationship between MARITAL STATUS, EDUCATION, and DEF_AMT")
plt.xlabel("MARRIAGE")
plt.ylabel("Count")

# Adjust the legend position
plt.legend(title="EDUCATION", loc="upper right")

plt.show()
print('The graph shows that the university students are most likely to default when they are single. However, both university students and highschool students are likely to default when they are married')
```
